```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://ks.wjx.top/jq/24386487.aspx'
headers = {'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/49.0.2623.221 Safari/537.36 SE 2.X MetaSr 1.0"}
postData = {'txtExt':'电子信息学院刘恒宇',
'txtPassword':'2018261681'}
urlText = requests.post(url, headers=headers, data=postData).text
soup = BeautifulSoup(urlText, 'lxml')
question = soup.find_all('div', class_='div_question')
```

```python
for item in question:
    print(item.text)
```
